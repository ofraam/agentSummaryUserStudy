```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

```

```python
df = pd.read_csv('results/Exp1Exp2_correctnessMerged.csv')
df['Comparison'] = df['compType'].apply(lambda x: 'Low vs. Medium' if x == 1 else ('Medium Vs. High' if x == 2 else 'Low Vs. High'))
df['Condition'] = df['condition'].apply(lambda x: 'First' if x == 'first' else ('Likelihood-Based' if x == 'random' else ('HIGHLIGHTS' if x == 'max' else 'HIGHLIGHTS-DIV')))

df['correct'].astype('int64')
df.head()
```

```python

sns.set(font_scale=1.5, style="whitegrid")
figure(figsize=(8, 6), dpi=80)

ax = sns.barplot('Comparison','correct',hue='Condition',data=df, hue_order=['First','Likelihood-Based','HIGHLIGHTS','HIGHLIGHTS-DIV'],palette='deep')
plt.legend(loc='center left', bbox_to_anchor=(1.01, 0.85), ncol=1)
# plt.show()
plt.savefig('results/highlightsExp12agg_all.png')
```
